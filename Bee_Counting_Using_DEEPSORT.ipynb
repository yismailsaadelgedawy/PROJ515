```python
import cv2
from ultralytics import YOLO

# Load the model
model = YOLO('/content/best.pt')

# Open the video file
video_path = '/content/bee_video.mp4'
output_path = 'output_video.mp4'

cap = cv2.VideoCapture(video_path)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

object_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            confidence = box.conf[0]
            class_id = box.cls[0]
            label = f'{class_id} {confidence:.2f}'

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

            object_count += 1

    out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()

print(f'Total objects counted: {object_count}')
```

```python
import cv2
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort
import numpy as np

# Load the YOLOv8n model
model = YOLO('/content/best.pt')

# Initialize DeepSORT
tracker = DeepSort(max_age=30, n_init=3, nms_max_overlap=1.0, max_cosine_distance=0.2)

# Open the trimmed video file
video_path = '/content/trimmed_video.mp4'
output_path = '/content/output_video_deepsort.mp4'

cap = cv2.VideoCapture(video_path)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

unique_track_ids = set()
frame_count = 0
track_id_to_info = {}

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    print(f'Processing frame {frame_count}')

    results = model(frame)
    detections = []

    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            confidence = box.conf[0]
            class_id = box.cls[0]
            width = x2 - x1
            height = y2 - y1
            detections.append(([x1, y1, width, height], confidence, class_id))

    # Update tracker with detections
    tracks = tracker.update_tracks(detections, frame=frame)

    for track in tracks:
        if not track.is_confirmed() or track.time_since_update > 1:
            continue
        x1, y1, x2, y2 = map(int, track.to_tlbr())
        track_id = track.track_id

        # Store class ID and confidence for the track ID
        if track_id not in track_id_to_info:
            for detection in detections:
                if detection[0] == [x1, y1, x2 - x1, y2 - y1]:
                    track_id_to_info[track_id] = (detection[2], detection[1])
                    break

        class_id, confidence = track_id_to_info.get(track_id, (-1, 0))
        label = f'ID: {track_id} Class: {class_id} Conf: {confidence:.2f}'

        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        unique_track_ids.add(track_id)

    out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()

print(f'Total unique objects tracked: {len(unique_track_ids)}')
```
